```python
#Importing Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the CSV file
df = pd.read_csv("hotel_bookings.csv")

# Show first few rows
df.head()
```

```python
# Checking shape of dataframe
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# Checking data types and missing values
df.info()

# Quick statistical summary
df.describe(include='all').T
```

```python
#Check for Missing Values
df.isnull().sum().sort_values(ascending=False)
```

**1.Cleaning Missing Values**

```python

# Fill 'children' with 0
df['children'].fillna(0, inplace=True)

# Fill 'agent' and 'company' with 0 (means "no agent/company")
df['agent'].fillna(0, inplace=True)
df['company'].fillna(0, inplace=True)

# Drop rows where 'country' is missing (very few)
df.dropna(subset=['country'], inplace=True)
```

```python
#Now confirming that everything is clean:
df.isnull().sum().sum()
```

**2.Remove Duplicate Rows**

```python
# Checking how many duplicate rows exist
print(f"Duplicate Rows: {df.duplicated().sum()}")

# Drop/Removing them
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

**3. Checking for Invalid Records** (e.g., 0 adults)
Logically, a booking should have at least one adult, child, or baby.

```python
# Bookings with 0 adults, 0 children, and 0 babies
df[(df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0)]
```

```python
#if such rows exist remove them using  ~ (tilde) operator: ~(...)This is the NOT operator in pandas. So it flips True â†’ False and False â†’ True.
df = df[~((df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0))]
```

**4. Creating a Proper Date Column**: Weâ€™ll combine arrival_date_year, arrival_date_month, and arrival_date_day_of_month into one clean arrival_date


```python
# Convert month name to number
df['arrival_date_month'] = pd.to_datetime(df['arrival_date_month'], format='%B').dt.month

# Create proper date column
df['arrival_date'] = pd.to_datetime(dict(
    year=df['arrival_date_year'],
    month=df['arrival_date_month'],
    day=df['arrival_date_day_of_month']
))
```

```python
#Previewing it
df[['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month', 'arrival_date']].head()
```

**5. Renaming Some Confusing Columns**

```python
df.rename(columns={
    'adr': 'avg_daily_rate',
    'stays_in_weekend_nights': 'weekend_nights',
    'stays_in_week_nights': 'week_nights'
}, inplace=True)
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (10,6)
```

## Exploratory Data Analysis and Visualizations

### **Question 1: What is the total number of bookings per hotel?**

```python
df['hotel'].value_counts()
```

```python
# Bar chart for visualization
sns.countplot(data=df, x='hotel', palette='pastel')
plt.title("Total Bookings by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Number of Bookings")
plt.show()
```

### **Q2: Whatâ€™s the cancellation rate for each hotel?**

```python
cancel_rate = df.groupby('hotel')['is_canceled'].mean() * 100
cancel_rate
```

```python
sns.barplot(x=cancel_rate.index, y=cancel_rate.values, palette='Set2')
plt.title("Cancellation Rate (%) by Hotel Type")
plt.ylabel("Cancellation Rate (%)")
plt.xlabel("Hotel")
plt.show()
```

### **Q3: Whatâ€™s the average stay duration by hotel type?**



```python
df['total_nights'] = df['weekend_nights'] + df['week_nights']
df.groupby('hotel')['total_nights'].mean()
```

```python
sns.boxplot(data=df, x='hotel', y='total_nights', palette='coolwarm')
plt.title("Average Stay Duration by Hotel")
plt.show()
```

### **Q4: What months see the most bookings?**


```python
df['arrival_date_month_name'] = df['arrival_date'].dt.month_name()
monthly = df['arrival_date_month_name'].value_counts().sort_index()
monthly
```

```python
# creating column chart to visualize trend
sns.countplot(data=df, x='arrival_date_month_name', order=monthly.index)
plt.title("Bookings per Month")
plt.xticks(rotation=45)
plt.show()
```

### **Q5: Which countries have the most guests?**

```python
top_countries = df['country'].value_counts().head(12)
top_countries
```

```python
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("Top 10 Guest Countries")
plt.xlabel("Number of Bookings")
plt.ylabel("Country")
plt.show()
```

### **Q6: How does lead time affect cancellations?**

```python
sns.boxplot(data=df, x='is_canceled', y='lead_time', palette='husl')
plt.title("Lead Time vs Cancellation")
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])
plt.ylabel("Lead Time (days)")
plt.show()
```

### **Q7: Trend of ADR (average daily rate) over months?**

```python
df['month'] = df['arrival_date'].dt.month
monthly_adr = df.groupby('month')['avg_daily_rate'].mean()

monthly_adr.plot(kind='line', marker='o')
plt.title("Monthly Average Daily Rate (ADR)")
plt.xlabel("Month")
plt.ylabel("ADR")
plt.grid(True)
plt.show()
```

### **Q8: Most frequently booked room types?**



```python
df['reserved_room_type'].value_counts()
```

```python
room_counts = df['reserved_room_type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    room_counts,
    labels=room_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.85,           # Pulls percentage closer to edge
    labeldistance=1.1,          # Moves labels outside
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}  # Adds separation
)
plt.title("Distribution of Reserved Room Types")
plt.tight_layout()
plt.show()
```

### **Q9: What market segments bring the most revenue?**

```python
revenue_by_segment = df.groupby('market_segment')['avg_daily_rate'].sum().sort_values(ascending=False)
sns.barplot(x=revenue_by_segment.values, y=revenue_by_segment.index, palette="mako")
plt.title("Revenue by Market Segment")
plt.xlabel("Total Revenue")
plt.show()
```

### **Q10: Distribution of special requests?**

```python
sns.countplot(data=df, x='total_of_special_requests', palette='flare')
plt.title("Total Special Requests by Guests")
plt.show()
```

### **Q11: Average number of guests per booking?**

```python
df['total_guests'] = df['adults'] + df['children'] + df['babies']
df['total_guests'].mean()
```

```python
sns.histplot(df['total_guests'], bins=20, kde=True)
plt.title("Distribution of Total Guests per Booking")
plt.show()
```

### **Q12: Do guests with children stay longer?**

### **Q13: Booking behavior in resort vs city hotel?**

```python
sns.histplot(data=df, x='lead_time', hue='hotel', element='step', stat='density', common_norm=False)
plt.title("Lead Time Comparison: City vs Resort Hotel")
plt.show()
```

### **Q14: Relationship between deposit type and cancellation?**

```python
sns.countplot(data=df, x='deposit_type', hue='is_canceled')
plt.title("Deposit Type vs Cancellation")
plt.xticks(rotation=0)
plt.legend(title="Cancelled", labels=["No", "Yes"])
plt.show()
```

### **Q15: Which distribution channels are most used?**

```python
df['distribution_channel'].value_counts()
```

```python
sns.countplot(data=df, x='distribution_channel', palette='tab10')
plt.title("Distribution Channels Used")
plt.xticks(rotation=0)
plt.show()
```

##  Key Insights & Business Recommendations

-  City hotels had higher cancellations (~40%) than resort hotels.
-  Bookings peaked in Julyâ€“August; low in Novemberâ€“December.
-  Most guests came from Portugal, followed by UK and France.
-  Higher lead time was strongly linked with cancellation.
-  Room Type A was most frequently booked (~65%).
-  Guests with children tended to stay longer.
-  Revenue was highest from the "Online TA" segment.

>  Hotel managers can reduce cancellations by targeting short lead-time customers and offering discounts during off-peak months.
